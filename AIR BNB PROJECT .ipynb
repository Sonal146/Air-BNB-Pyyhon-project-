# What is the Relation Between Price and Room Type 

```python
# Example: Remove '$' and ',' characters if they exist
# Make sure to adjust this based on what non-numeric characters are actually in your 'price' column
df['price'] = df['price'].astype(str).str.replace(r'[$,]', '', regex=True) # Convert to string first, then remove $ and ,

# Convert the cleaned column to numeric (float is usually safe for prices)
# errors='coerce' will turn any values that still can't be converted into NaN (Not a Number)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Optional: Check for and handle any NaN values created during conversion
print(f"Number of rows with invalid price data (NaN): {df['price'].isnull().sum()}")
# You might want to drop rows where price couldn't be converted:
# df.dropna(subset=['price'], inplace=True)
```

# How Has The Number of Review Change Over The Time ?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it has a 'last review' column

# Convert 'last review' column to datetime objects
df['last review'] = pd.to_datetime(df['last review'])

# Group by month and count the number of reviews
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

# Create the plot
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='hotpink')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
